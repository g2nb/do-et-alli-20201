## Imports and definitions 

```python
# Imports and definitions

import pandas as pd
from tqdm.notebook import tqdm
import pickle
from urllib.request import urlopen
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
def logplus1(x):
    return np.log2(x+1)

fzd7 = dict(size=10,color='black')
other = dict(size=5,color='black')
mito = dict(size=5,color='green')

def annotate(ax,gene,df,color,kind='right',start_index = 1):
    # this function adds a marker (x or +) to a figure indicating the expression of a gene of interest
    x=start_index
    for i in df.loc[gene]:
        if kind=='right':
            ax.annotate(f'x', xy=(x, i), xytext=(x, i),va='center',ha='center',**color)
        elif kind=='left':
            ax.annotate(f'+', xy=(x, i), xytext=(x, i),va='center',ha='center',**color)
        x+=1
titlefont = {'fontname':'Arial','fontsize':'16'}
axisfont = {'fontname':'Arial','fontsize':'12'}
```

## Load and lightly preprocess data

<div class="well well-sm">  
    
Data was originally from: Rahman M, Jackson LK, Johnson WE, Li DY, Bild AH, Piccolo SR. Alternative preprocessing of RNA-Sequencing data in The Cancer Genome Atlas leads to improved analysis results. Bioinformatics. 2015;31(22):3666-3672. [doi:10.1093/bioinformatics/btv377](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC4804769/)

```python
## I've commented out these lines because they require a large amount of memory. If a user wants to start from raw data they should uncomment this and let the GenePattern Notebook admins know the need for a large memory node.
## Most people can start from the file below which is simply a pickle/serialized form of the same DataFrame created in this cell

# tcga_tpm = pd.read_table('https://datasets.genepattern.org/data/publications/Do_2021/rahman_tcga_tpm.txt',sep='\t')
# tcga_tpm = tcga_tpm.set_index('Unnamed: 0')
# tcga_tpm.index.name='Gene'
# tcga_tpm.head()
```

```python
# Load and lightly preprocess data
tcga_tpm = pickle.load(urlopen('https://datasets.genepattern.org/data/publications/Do_2021/tcga_tpm.p'))
display(tcga_tpm.head())
tumor_types = pd.read_table('https://datasets.genepattern.org/data/publications/Do_2021/rahman_tcga_cancertype_samples.txt',index_col=0,header=None)
tumor_types.index.name = 'ID'
tumor_types['type']=tumor_types[1]
tumor_types = tumor_types.drop(1,axis=1)
display(tumor_types.head())
```

## Separating data into different cancer types

```python
#Separating data into different cancer types
print('Working with each cancer type from TCGA')
tcga_medians = pd.DataFrame()
for cancer_type in tqdm(tumor_types['type'].unique()):
#     print('Working with',cancer_type)
    ids = list(tumor_types.index[tumor_types['type'] == cancer_type])
    
    temp_primary = tcga_tpm[[idx for idx in ids if idx[13:15]=='01']]
    temp_normal = tcga_tpm[[idx for idx in ids if idx[13:15]=='11']]
    
    if len(temp_primary.columns)!=0:
        tcga_medians[f'{cancer_type}'] = temp_primary.median(axis=1)
    if len(temp_normal.columns)!=0:
        print('This should not be happening')

tcga_medians = tcga_medians[sorted(tcga_medians.columns)]
tcga_medians_log = tcga_medians.apply(logplus1)
tcga_medians.head()
```

## Creating figure

```python
fig, ax = plt.subplots(figsize=(8,3.5),dpi=300)
ax = sns.boxplot(data=tcga_medians_log,ax=ax)

annotate(ax,'FZD7',color=fzd7,df=tcga_medians_log,start_index = 0)

plt.gca().set_title('TCGA',**titlefont)
plt.gca().set_ylabel('Median Expression of Genes\nlog2(median(TPM)+1)',**axisfont)
plt.xticks(rotation=90,**axisfont);
plt.yticks(**axisfont);
```

# Export cancer types with high FZD7 expression

We will use the classification from TCGAbiolinks in future analyses, so we prepare those dataframes here

```python
subtypes_from_TCGAbiolinks = pd.read_excel('subtypes_from_TCGAbiolinks_preprocessed.xlsx',index_col=0)
subtypes_from_TCGAbiolinks.head()
```

<div class="alert alert-warning">
<p class="lead"> Warning <i class="fa fa-exclamation-triangle"></i></p>
Only run this cell if you want to replicate every intermediate step of the paper. This step takes about 2 minutes to complete but creates about 1GB of unnecesary data.
    
The files created on this step, have been uploaded to https://datasets.genepattern.org/?prefix=data/publications/Do_2021/
</div>

```python
# subtypes_from_TCGAbiolinks['Cancer_Type'].unique()
for cancer_type in tqdm(['BRCA','GBM','LUSC','OV','UCS']):
    print('Working with',cancer_type)
    
    ids = list(tumor_types.index[tumor_types['type'] == cancer_type])
    temp_primary = tcga_tpm[[idx for idx in ids if idx[13:15]=='01']]
    temp_primary.columns = [i[:12] for i in temp_primary.columns]
    print(f'There are {len(temp_primary.columns)} samples.')
    dupes = temp_primary.columns[temp_primary.columns.duplicated()].unique()
    print(f'There are {len(dupes)} duplicated samples. Clening them up...',end = '')
    
    #The duplicated columns represent multiple vials or portions from the same patient & sample,
    #so I am taking the median expression of those
    #https://docs.gdc.cancer.gov/Encyclopedia/pages/images/TCGA-TCGAbarcode-080518-1750-4378.pdf

    for what in dupes:
        temp_median = temp_primary[what].median(axis=1)
        temp_primary = temp_primary.drop(what,axis=1)
        temp_primary[what] = temp_median
    print(f' done! There are {len(temp_primary.columns)} samples now.')
    
    # Two cancer types have slighly different names in the different publications
    if cancer_type == 'GBM':
        cancer_type = 'GBM_LGG'
    if cancer_type == 'OV':
        cancer_type = 'OVCA'
        
    cancer_subtypes = subtypes_from_TCGAbiolinks[subtypes_from_TCGAbiolinks['Cancer_Type']==cancer_type]
    print(f"There are {len(cancer_subtypes)} samples in TCGAbiolinks.")
    common = list(set(cancer_subtypes.index)&set(temp_primary.columns))
    print(f'{len(common)} samples in common')
    temp_primary = temp_primary[common]
    print('Saving dataset...',end = '')
    temp_primary.to_csv(f'{cancer_type}_TPM.csv')
    cancer_subtypes.loc[common].to_excel(f'{cancer_type}_subtypes.xlsx')
    print(' done!')
    print('======================================')
print('All done!')
```
